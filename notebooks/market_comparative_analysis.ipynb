# Price Pattern Analysis: AAPL vs NIFTY 50

## AAPL (Apple Stock)

  - There are periodic pullbacks followed by recovery.

- **Increasing Volatility Over Time**
  - Early period: relatively stable movements.
  - Later period: larger fluctuations and sharper price swings.

- **Volatility Clustering**
  - Periods of calm are followed by calm.
  - Periods with large fluctuations appear grouped together.

- **Sharp Drops and Recoveries**
  - Sudden drops (likely due to market events) followed by quick recovery.
  - Indicates resilience of large-cap stocks.

## NIFTY 50 (Index)

### Key Observations

- **Steady Upward Trend**
  - NIFTY also shows long-term growth but smoother compared to AAPL.
  - Reflects diversified market behavior.

- **Lower Relative Volatility**
  - Compared to AAPL, fluctuations are more controlled.
  - Index reduces company-specific risk.

- **Major Market Shock Visible**
  - A sharp drop is observed (around mid timeline).
  - Likely corresponds to a macro event (e.g., COVID crash).

- **Gradual Recovery After Crash**
  - Post-crash recovery is steady and sustained.
  - Shows strength of broader market.

- **Volatility Clustering Present**
  - High volatility periods occur during crises.
  - Calm periods dominate during stable market conditions.

## Comparative Insights

- **AAPL vs NIFTY**
  - AAPL → Higher volatility, sharper movements.
  - NIFTY → Smoother, more stable due to diversification.

- **Risk Characteristics**
  - Individual stock (AAPL): higher risk, higher fluctuations.
  - Index (NIFTY): lower risk, more stable growth.

- **Volatility Behavior**
  - Both show volatility clustering, which justifies using GARCH models.

- **Non-Stationarity**
  - Both price series show clear trends.
  - Confirms that prices are non-stationary → need to use returns.

## Conclusion

- Both assets exhibit:
  - upward trend  
  - volatility clustering  
  - structural changes  

- These characteristics make them suitable for:
  - time series modeling  
  - volatility forecasting

# Log Return Analysis: AAPL vs NIFTY 50

## AAPL Log Returns

### Key Observations

- **Mean Around Zero**
  - Returns fluctuate around zero.
  - Indicates no consistent upward or downward drift in returns.

- **Volatility Clustering**
  - Periods of small fluctuations are followed by small fluctuations.
  - Periods of large spikes occur in clusters.
  - This is a key property of financial time series.

- **High Volatility Periods**
  - Around the middle of the timeline, large spikes are observed.
  - Indicates market stress or major events.

- **Presence of Outliers**
  - Extreme positive and negative spikes are visible.
  - Suggests heavy tails in distribution.

- **Time-Varying Variance**
  - Variance is not constant over time.
  - Confirms heteroskedastic behavior.

## NIFTY 50 Log Returns

### Key Observations

- **Mean Around Zero**
  - Similar to AAPL, returns are centered around zero.

- **Lower Volatility Compared to AAPL**
  - Fluctuations are generally smaller.
  - Reflects diversified index behavior.

- **Clear Volatility Clustering**
  - High volatility concentrated around specific periods.
  - Especially visible during major market events.

- **Extreme Negative Spike**
  - A sharp drop is visible (likely market crash event).
  - Followed by increased volatility.

- **More Stable Behavior Overall**
  - Compared to AAPL, movements are smoother.

## Comparative Insights

- **Volatility Level**
  - AAPL → higher volatility  
  - NIFTY → lower volatility  

- **Clustering Behavior**
  - Both exhibit strong volatility clustering.
  - This supports the use of ARCH/GARCH models.

- **Non-Constant Variance**
  - Variance changes over time in both series.
  - Violates constant variance assumption of simple models.

- **Heavy Tails**
  - Presence of extreme spikes suggests non-normal distribution.

## Conclusion

- Returns are:
  - centered around zero  
  - heteroskedastic  
  - clustered in volatility  

- These characteristics confirm:
  → Financial returns are suitable for volatility modeling  
  → ARCH/GARCH models are appropriate

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(aapl["log_return"], lags=40)
plt.title("ACF of AAPL Log Returns")
plt.show()
```

# ACF of AAPL Log Returns: Key Takeaways

- Most autocorrelation values are **close to zero**  
- Almost all lags lie within the **confidence band**  
- Indicates **no significant linear dependence** in returns  

## Interpretation

- Returns behave like **white noise**  
- Past returns do **not predict future returns**  
- Weak or no serial correlation in mean  

## Important Insight

- Even though returns show no autocorrelation:  
  → volatility may still be autocorrelated  
- This is why we check:
  → ACF of **squared returns** next  

## Conclusion

- Mean process is weak → ARMA may be simple or unnecessary  
- Focus should shift to **volatility modeling (GARCH)**  

```python
plot_acf(nifty50["log_return"], lags=40)
plt.title("ACF of Nifty50 Log Returns")
plt.show()
```

# ACF of NIFTY 50 Log Returns: Key Takeaways

- Autocorrelation values are **very close to zero**  
- Almost all lags fall within the **confidence interval**  
- No strong or consistent pattern across lags  

## Interpretation

- Returns behave like **white noise**  
- Past returns have **little to no predictive power**  
- No significant linear dependence in mean  

## Important Insight

- Similar to AAPL:
  → returns are not autocorrelated  
  → but volatility may still be  

- Next step:
  → check **ACF of squared returns** for volatility clustering  

## Conclusion

- Mean process is weak  
- Simple or no AR component is sufficient  
- Focus should be on **volatility modeling (GARCH)**  

```python
plot_acf(aapl["log_return"]**2, lags=40)
plt.title("ACF of AAPL Squared Log Returns")
plt.show()
```

# ACF of AAPL Squared Log Returns: Key Takeaways

- Significant positive autocorrelation at multiple lags  
- Many lags lie **outside the confidence band**  
- Autocorrelation decays slowly over time  

## Interpretation

- Strong evidence of **volatility clustering**  
- Large returns tend to be followed by large returns  
- Small returns tend to be followed by small returns  

## Important Insight

- Even though raw returns had no autocorrelation:  
  → squared returns show clear dependence  
- Indicates **conditional heteroskedasticity**  

## Conclusion

- ARCH effects are present  
- GARCH type models are **appropriate**  
- Volatility is **predictable over time**, even if returns are not  

```python
plot_acf(nifty50["log_return"]**2, lags=40)
plt.title("ACF of Nifty50 Squared Log Returns")
plt.show()
```

# ACF of NIFTY 50 Squared Log Returns: Key Takeaways

- Significant positive autocorrelation at early lags  
- Several lags lie **outside the confidence band**  
- Autocorrelation gradually decreases over time  

## Interpretation

- High volatility periods are followed by high volatility  
- Low volatility periods are followed by low volatility  

## Important Insight

- Similar to AAPL:
  → raw returns show no autocorrelation  
  → squared returns show strong dependence  

- Confirms **conditional heteroskedasticity**  

## Conclusion

- ARCH effects are present  
- GARCH models are **suitable**  
- Volatility is **time dependent and predictable**
